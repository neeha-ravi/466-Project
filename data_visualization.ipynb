```python
from data.data_processing import *
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

data  = get_data()



df = pd.DataFrame(data)

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(data.head())

# Check for missing values
print("\nMissing values:")
print(data.isnull().sum())

print("\nSummary statistics:")
print(data.describe())

# Plotting the distribution of the target variable
plt.figure(figsize=(40, 30))
sns.countplot(x='Performance Index', data=df)
plt.title('Distribution of the target variable')
plt.show()

# Explore the distribution of numerical features using histograms
plt.figure(figsize=(30, 20))
for i, col in enumerate(df.columns[:-1], 1):
    plt.subplot(4, 4, i)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()

# Visualize relationships between features and target variable (species)
plt.figure(figsize=(100, 100))
for i, col in enumerate(data.columns[:-1], 1):
    plt.subplot(4, 4, i)
    sns.boxplot(data=data, x='Performance Index', y=col)
    plt.title(f"{col} by Performance Index")
plt.tight_layout()
plt.show()




# Encode categorical variables using LabelEncoder
label_encoder = LabelEncoder()
categorical_columns = ['Extracurricular Activities']  # 
for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col])


# Correlation matrix as heatmap
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()


```
